```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
import statsmodels.api as sm

color = sns.color_palette()
plt.style.use('ggplot')
plt.rcParams['font.sans-serif']=['SimHei'] 
```

```python
data = pd.read_excel('../table/体育成绩.xlsx')
data
```

```python
data.iloc[:,1]
```

```python
x = range(0,15)
colnm = data.columns.tolist()
plt.figure(figsize = (15, 10))
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.plot(x,data.iloc[:,i-1],'o')
    plt.title(colnm[i-1])
    plt.tight_layout()
    # title
```

```python
# 皮尔逊相关系数
data.corr()
```

```python
# 肯德尔相关系数
data.corr('kendall')
```

```python
# 斯皮尔曼相关系数
data.corr('spearman')
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(data=data.corr(),annot=True,cmap="RdBu_r")
```
